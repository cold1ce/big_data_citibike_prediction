# Historical Weather Data Processing

##### Transforming a Dataset containing weather information of several locations in the united states to fit it to our utilization table

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import pyplot
temperature = pd.read_csv("C:/Users/Julian/Desktop/BigData/historical-hourly-weather-data/temperature.csv")
```

```python
temperature.head(2)
temperature['datetime'] = pd.to_datetime(temperature["datetime"])
temperature.dtypes
```

```python
temperature_ny = temperature[["datetime", "New York"]]
```

```python
temperature_ny.head(5)
```

```python
# Temperature is given in Kelvin, we want to use celsius
# Formula: Celsius = Kelvin - 273.15
temperature_ny.loc[:,"temperature"] = (temperature_ny["New York"] - 273.15)
temperature_ny.head(100)
```

```python
temperature_ny = temperature_ny[["datetime", "temperature"]]
temperature_ny.head(5)
```

```python
# Make outliners visible via 
temperature_ny = temperature_ny.fillna(value=-100.0)

temperature_ny.plot(x="datetime",y="temperature")
```
